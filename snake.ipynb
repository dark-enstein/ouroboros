```python
# pipeline

# create or retrieve dataset embedding
# get user features and actions
# explore in-network and out-network posts
# hydrate retrieved posts
# pre-filtering (user guardrails)
# scoring:
# post-filtering (social guardrails)
```

```python
from datasets.retrieve import Embedding
from context.user.user import User
from core.utils import filter, rank

data_src = "datasets/feat.csv"

# create or retrieve dataset embedding
embed = Embedding(data_src)

# get user features and actions
user_state = User.embedding()

# explore in-network and out-network posts
in_network = embed.get_innetwork_posts(user_state) * 0.6
out_network = embed.get_outnetwork_posts(user_state) * 0.4

# hydrate retrieved posts
in_network_hydrate = [i.hydrate() for i in in_network.list]
out_network_hydrate = [i.hydrate() for i in out_network.list]
hydrated_posts = in_network_hydrate + out_network_hydrate

# prefiltering
prefiltered_results = User.filter(hydrated_posts)

# scoring based on posts current engagement
engagement_based_ranking = rank(prefiltered_results)

# post-filtering (social guardrails)
final_filter = filter(engagement_based_ranking)

# display to user
# -> UI
# <- record engagement params
engagement_params = None

# record engagement to the filter scaffold in realtime: databaxse ops
User.update_filters(engagement_params)
```
